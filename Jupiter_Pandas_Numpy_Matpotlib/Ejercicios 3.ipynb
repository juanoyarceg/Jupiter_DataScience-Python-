```python
# Desafío 6#

import pandas as pd
import numpy as np

df=pd.read_csv("flights.csv")
media_pas=df["passengers"].mean()
ds_pas=df["passengers"].std()

df["outliers"]=0 
cont=0
for indice,fila in df.iterrows():
    if fila['passengers']< (media_pas-ds_pas) or fila['passengers'] >(media_pas+ds_pas) :
        df.at[indice,'outliers'] =1
        cont+=1
        
print("El numero de registros como casos extremos es: ", cont) #El modo más sencillo es con un contador#
data_filter = df[df['outliers'] == 1] 
print ("El numero de registros como casos extremos es: ", data_filter['outliers'].count())#otra forma con count#
df.set_index(["outliers", "passengers"]).count(level="outliers")#Otra forma es usando set_index #
```
